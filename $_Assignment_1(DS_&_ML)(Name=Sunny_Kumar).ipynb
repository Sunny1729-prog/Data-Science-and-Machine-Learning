```python
import numpy as np

M = int(input("Enter value M: "))
N = int(input("Enter value N: "))
A = int(input("Enter value A: "))

np.random.seed(0)
x=np.random.random((M,N))
y=np.random.random((N,A))

print("\nFirst array:\n",x,end="\n\n")
print("Second array:\n",y,end="\n\n")
print("Dot product of two arrays:\n",x@y)
```

```python
import numpy as np

items=["Sunny","Kumar","ABC",6832,"abc",368]
array1=np.array(items)
print("Original array:\n",items,end='\n\n')
print("Digits only:\n",np.char.isdigit(array1),end='\n\n')
print("Lower cases only:\n",np.char.islower(array1),end='\n\n')
print("Upper cases only:\n",np.char.isupper(array1),end='\n\n')
```

```python
import numpy as np

(x,y)=map(int,input().split())     #Method to take space seperated input and map them to a data structure

lst1=list(np.arange(1,(16*x),x))
lst2=list(np.arange(1,(16*y),y))

np1=np.array(lst1).reshape(4,4)
np2=np.array(lst2).reshape(4,4)

np3=np1-np2
print(np3.ravel())
```
