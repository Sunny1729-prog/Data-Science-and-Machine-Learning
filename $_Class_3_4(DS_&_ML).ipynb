```python
# Using random.randint

np.random.seed(9999)
g=np.random.randint(low=0,high=50,size=(5,5))
h=np.random.randint(low=0,high=100,size=(5,5))
print(g)
print('\n')
print(h)
```

```python
# Using mathematiacal operations on arrays

print(g+h,end="\n\n")
print(g-h,end="\n\n")
print(g*h,end="\n\n")
print(g/h,end="\n\n")

```

```python
# Using statistical operations
# axis 0 --> column
# axis 1 --> row
#[[22 49 47 45 28]
#[37 25 38 24 16]
#[12 40 33 46  5]
#[33 32 15 38 36]
#[20 12 22  8 28]]

print(np.sum(g),end="\n\n")          #sum of all elements in the array
print(np.sum(g,axis=0),end="\n\n")   #sum of all elements in the array column wise
print(np.sum(g,axis=1),end="\n\n")   #sum of all elements in the array row wise
print(np.max(g),end="\n\n")          #max of all elements in the array
print(np.max(g,axis=0),end="\n\n")   #max of all elements in each column
print(np.max(g,axis=1),end="\n\n")   #max of all elements in each row
print(np.min(g),end="\n\n")          #min of all elements in the array
print(np.min(g,axis=0),end="\n\n")   #min of all elements in each column
print(np.min(g,axis=1),end="\n\n")   #min of all elements in each row
print(np.transpose(g),end="\n\n")    #transpose of the given matix
```

```python
# vstack(no of columns should match) and hstack(no of rows should match)
i=np.hstack([g,h])
j=np.vstack([g,h])
print(i,end="\n\n")
print(j)
```

###Two cool features of Python NumPy: Mutating by slicing and Broadcasting

image.png

```python
# Example code to demonstrate mutation by slicing

mat = np.array([[11,12,13],[21,22,23],[31,32,33]])
print("Original matrix")
print(mat)
mat_slice = mat[:2,:2] # Simple indexing
print ("\nSliced matrix")
print(mat_slice)
print ("\nChange the sliced matrix")
mat_slice[0,0] = 1000
print (mat_slice)
print("\nBut the original matrix? WHOA! It got changed too!")
print(mat)
```

```python
# To stop the above thing from happening, use array method of numpy

mat = np.array([[11,12,13],[21,22,23],[31,32,33]])
print("Original matrix")
print(mat)
mat_slice = np.array(mat[:2,:2]) # Notice the np.array method
print ("\nSliced matrix")
print(mat_slice)
print ("\nChange the sliced matrix")
mat_slice[0,0] = 1000
print (mat_slice)
print("\nBut the original matrix? NO CHANGE this time:)")
print(mat)
```

```python
#Creation of a start matrix
start = np.zeros((4,3))
print(start)
```

```python
# create a rank 1 ndarray with 3 values
add_rows = np.array([1, 0, 2])
print(add_rows)
```

```python
# 1x3 vector is duplicated 4 times to match the row dimension of the zero matrix 
# and those values are added to the 4x3 matrix.

y = start + add_rows  # add to each row of 'start'
print(y)
```

```python
# create an ndarray which is 4 x 1 to broadcast across columns
add_cols = np.array([[0,1,2,3]])
add_cols = add_cols.T
print(add_cols)
```

```python
# add to each column of 'start' using broadcasting
y = start + add_cols 
print(y)
```

```python
# a scalar is treated like a 1x1 matrix and duplicated exactly to the size of 
# the additive matrix to execute the operation.

# this will just broadcast in both dimensions
add_scalar = np.array([100])  
print(start+add_scalar)
```
