```python
df1=pd.DataFrame({'Student_name':['Sam','Jordan'],'Maths':[88,89],'Physics':[91,92]})
df2=pd.DataFrame({'Student_name':['Michael','Gill'],'Maths':[88,85],'Chemistry':[98,96]})
```

```python
display(df1)
display(df2)
```

```python
# MERGE
# NaN denotes 'Not a Number'
df3=pd.merge(df1,df2,on='Student_name',how='outer')
df3
```

```python
df4=pd.merge(df1,df2,on=['Student_name','Maths'],how='outer')
df4
```

```python
df5=df4[['Student_name','Physics']][not(pd.isna(df4['Physics']))] #selecting columns and applying condition at same time
df5
```

```python
df4.rename(columns={'Student_name':'Student'},inplace=True)
df4
```
