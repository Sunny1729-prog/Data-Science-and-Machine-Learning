###Line plot

```python
import numpy as np
import pandas as pd
```

```python
!wget https://raw.githubusercontent.com/nachikethmurthy/Source-Code-Dataset-for-Machine-Learning-using-Python/main/Data/Car_sales.csv
```

```python
df=pd.read_csv("/content/Car_sales.csv")
df.head()
```

```python
df = df.sort_values('Sales in thousands')
df.tail()
```

```python
plt.plot(df['Sales in thousands'])
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
x=[1,2,3,4,5,6,7,8,9]
x.sort(reverse=True)
y=[1,2,3,4,5,6,7,8,9]
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x)
plt.xlim(0,10)
plt.ylim(-2,2)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title(" Sine wave vs Cosine wave")
plt.plot(x,y,":r",label="sine wave")
plt.plot(x,z,"--g",label="cosine wave")  #color='g', linestyle='--'
plt.show()
```

###Scatter plot

```python
# Scatter plot 
plt.plot(df['Fuel capacity'],df["Fuel efficiency"],"v")
plt.xlabel("Fuel capacity")
plt.ylabel("Fuel efficiency")
plt.show()
```

```python
plt.scatter(df['Fuel capacity'],df["Fuel efficiency"],marker="v", s=100) # 's' denotes size and 'marker' means type of marker
plt.xlabel("Fuel capacity")
plt.ylabel("Fuel efficiency")
plt.show()
```

###Bar chart

```python
result = df['Vehicle type'].value_counts()
result
```

```python
result.index
```

```python
result.values
```

```python
plt.bar(result.index, result.values)
plt.show()
```

```python
plt.barh(result.index, result.values)
plt.show()
```

###Pie chart

```python
plt.pie(result.values,labels=result.index,autopct = "%0.2f%%")
plt.show()
```

```python
#Altering the figure size
plt.figure(figsize=(5,5))
plt.pie(result.values,labels=result.index,autopct = "%0.2f%%")
plt.show()
```

```python
#Multiple plots in the same figure

x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(10,5))

plt.subplot(2,2,1) # no_of_row, no_of_col,index
plt.plot(x,y)

plt.subplot(2,2,2) 
plt.plot(x,z)

plt.subplot(2,2,3)
plt.plot(x,y)

plt.subplot(2,2,4) 
plt.plot(x,z)

plt.show()
```
