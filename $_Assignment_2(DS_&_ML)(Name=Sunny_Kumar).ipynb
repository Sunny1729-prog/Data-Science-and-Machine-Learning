```python
#Given dictionary
Subjects={'English' : {'Sam' : 60,'Jackson' : 74,'Ahree' : 85}, 
       'History' : {'Gloria' : 83,'Sam' : 65,'Isla' : 78,'Aron' : 72,'Gray' : 61}, 
       'Geography' : {'Jackson' : 92,'Gloria' : 95,'Isla' : 82,'Aron' : 75,'Ahree' : 76}, 
       'Mathematics' : {'Sam' : 99,'Gloria' : 74,'Jackson' : 89,'Ahree' : 85,'Gray' : 95}, 
       'Science' : {'Sam' : 89,'Aron' : 82,'Gray' : 78,'Isla' : 93,'Ahree' : 87} 
}
```

```python
#Creating a pandas series from given dictionary
ser=pd.Series(Subjects)
ser
```

```python
#Creating a dataframe from the given pandas series using apply() method
df3=ser.apply(pd.Series)
df3
```

```python
#Replacing null values with zeroes
df3=df3.fillna(0)
df3
```

```python
#Taking transpose
df3=df3.T
df3
```

```python
#Finding Average score of a student in all subjects
df3['Average']=df3.mean(axis=1)
df3
```

```python
data=pd.read_csv("cereal.csv")
data
```

```python
#Taking mean of ratings of different brands by grouping them using mfr
data2=data.groupby(['mfr'])['rating'].mean().sort_values(ascending=False)
data2
```

```python
#Ploting the ratings of various cereal manufactures
plt.bar(data2.index, data2.values, color='r')
plt.show()
```

```python
#Specifying the xtick range as 0-100 and inverting the axes
plt.barh(data2.index, data2.values)
plt.xlim(0,100)
plt.show()
```

```python
#Changing style of the graph to seaborn
plt.style.use('seaborn')
plt.barh(data2.index, data2.values)
plt.xlim(0,100)
plt.show()
```
