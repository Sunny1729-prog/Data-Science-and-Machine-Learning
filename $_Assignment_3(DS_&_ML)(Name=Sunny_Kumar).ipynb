```python
#Problem 1

total=200
num=int(input("Enter the number of satisfied customers: "))
prob=num/total

#Calculation of critical value
critical_value = stats.norm.ppf(0.90)

#Calulation of standard error of statistics
standard_error = sqrt((prob*(1-prob))/total)

marginal_error=critical_value*standard_error

print("Confidence interval: \n{:.5f} {:.5f}".format(prob-marginal_error,prob+marginal_error))
```

```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df.head()
```

```python
df['target'] = wine.target
df.tail()
```

```python
df2=df.sample(n=50,random_state=100)
df2
```

```python
#Calculation of critical value
critical_value = stats.norm.ppf(0.95)
critical_value
```

```python
#Calulation of marginal error
marginal_error=critical_value*(np.std(df['target'])/sqrt(len(df2.index)))
marginal_error
```

```python
print("Confidence interval: \n{:.5f} {:.5f}".format(df2['alcohol'].mean()-marginal_error,df2['alcohol'].mean()+marginal_error))
```
